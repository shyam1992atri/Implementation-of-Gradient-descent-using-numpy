```python
# y = mx + b
# m is slope, b is y-intercept
'''
this function will accumilate the the total error

'''
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))
```

```python
'''
this function will compute the gradient by taking account of all the points

'''
def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]
```

```python
'''
this functiion will compute all the gradiant descent for set number of iterations

'''
def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
          b, m = step_gradient(b, m, np.array(points), learning_rate)
    return [b, m]
```

```python
# this is a data frame to hold the variables of our interest to which 
# we will apply the gradient descent algorithm
pt=pd.DataFrame(index=[df['sacral_slope'],df['pelvic_incidence']])
```

```python
# the data frame is converted to a csv file
pt.to_csv('points.csv',header=False)
```

```python
'''
this is the function that will perform the
gradiant descent operation on the variables of our interest

'''
def run():
    points = np.genfromtxt("points.csv", delimiter=",")
    learning_rate = 0.0001
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = 1000
    print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print ("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    return [b,m]
```

```python
'''
this section is for 
post processing of results

'''

if __name__ == '__main__':
    val=run()
    
    x=df['sacral_slope'].tolist()
    y=df['pelvic_incidence'].tolist()
    plt.scatter(x,y)
    x_pred=np.linspace(0,130,1000)
    y_pred=val[0]+val[1]*x_pred
    plt.plot(x_pred,y_pred,'g')
    
    
```

```python

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
ax.scatter(x,y)
ax.plot(x_pred,y_pred,'g')
ax.set_title('plot of pelvic incidence vs sacral slope')
ax.set_xlabel('pelvic incidence')
ax.set_ylabel('pelvic incidence')
ax.legend(['gradiant descent prediction','original data'])
```
